Manipulating a Data Frame

```python
# adding a column in a data frame (the number of values should be same)
bc_df['Cancerous'] = bc_dataset.target
```

```python
bc_df.head()
```

```python
# removing a specific row
bc_df.drop(index=0, axis=0) # index is row number and axis=0 is row while axis=1 is column
# now index starts from 1 as 0th index was removed.
# if index=4, then the output index will be 0 1 2 3 5
```

```python
# removing a specifc column
bc_df.drop(columns='mean radius', axis=1)
```

```python
# locating a specific row based on index value
bc_df.iloc[4]
```

```python
# locating a specific column
print(bc_df.iloc[:,0] )# first col
print(bc_df.iloc[:,1] )# second col
print(bc_df.iloc[:,2] )# third col
print(bc_df.iloc[:,-1])# last col
```

Correlation

1. Postivite correlation (if one value increases/decreases upon the increase/decrease of another value)
2. Negative correlation (if one value increases/decreases upon the decrease/increase of another value)

corr 1 means perfect postivite correlation (for all possible values)

corr -1 means perfect negative correlation (for all possible values)

corr 0 means no significant relation

```python
# correlation
bc_df.corr()
```

```python
# correlation between specific columns
bc_df['mean radius'].corr(bc_df['mean area'])
```
