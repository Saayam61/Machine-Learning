Matplotlib:

Useful for making plots

```python
# importing matplotlib library (pyplot is enough for standard plotting, whole library is only used ocassionally)
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # 100 evenly spaced values from 0 to 100
y = np.sin(x)
z = np.cos(x)
```

```python
print('x = ', x)
```

```python
print('y = ', y)
```

```python
print('z = ', z)
```

Plotting the data

```python
# sine wave
plt.plot(x, y) # default plot is line plot so not specifying means default
plt.show() # shows the plot
```

```python
# cosine wave
plt.plot(x, z)
plt.show()
```

```python
# adding title, x-axis and y-axis labels
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10, 10, 20)
y = x ** 2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'b--')
plt.show()
```

Bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1]) # defines a rectangle from origin to 1, 1 where our bar will be plotted
languages = ['English', 'Japanese', 'French', 'Spanish', 'Chinese']
people = [70, 35, 50, 95, 170]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NO. OF PEOPLE')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ['English', 'Japanese', 'French', 'Spanish', 'Chinese']
people = [70, 35, 50, 95, 170]
ax.pie(people, labels = languages, autopct='%1.2f%%') # '%1.2f%%' defines as two decimal in % value of pie chart. can be any.
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color = 'g')
ax.scatter(x, z, color = 'b')
plt.show()
```

3D scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c = z, cmap = 'Blues')
# In x,y,z case 3 points x,y,z will be plotted but in x,y,s=z case only x,y will be plotted and z denotes the size.
plt.show()
```
