```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# applies default themes and settings to the plots like grids, pallete, scale, and context
sns.set()
```

Plotting a Vector

```python
# The plt.quiver function plots vectors as arrows, with their tails at specified coordinates and their directions
# and magnitudes defined by the input data.
plt.quiver(0, 0, 4, 5)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, scale_units = 'xy', angles = 'xy', scale = 1)
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.quiver(0, 0, -3, -6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Adding two vectors

```python
vector1 = np.asarray([0, 0, 2, 3])
vector2 = np.asarray([0, 0, 3, -2])
sum = vector1 + vector2
print(sum)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.quiver(0, 0, 3, -2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0, 0, 5, 1, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Subtraction of two vectors

```python
vector1 = np.asarray([0, 0, 2, 3])
vector2 = np.asarray([0, 0, 3, -2])
difference = vector1 - vector2
print(difference)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.quiver(0, 0, 3, -2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0, 0, -1, 5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Multiplying a vector by a scalar: The result will also be a vector

```python
vector1 = np.asarray([0, 0, 2, 3])
mul_by_scalar = 2 * vector1
print(mul_by_scalar)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.quiver(0, 0, 4, 6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
vector1 = np.asarray([0, 0, 2, 3])
mul_by_scalar = -0.5 * vector1
print(mul_by_scalar)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.quiver(0, 0, -1, -1.5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```
