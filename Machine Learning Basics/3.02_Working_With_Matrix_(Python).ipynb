```python
import numpy as np
```

Creating a Matrix

```python
matrix1 = np.array([[1, 2], [3, 4]])
print(matrix1)
```

```python
matrix1.shape
```

```python
matrix2 = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print(matrix2)
```

Creating a matrix with random values

```python
# random float matrix (always between 0 and 1)
random_float_matrix = np.random.rand(3, 3)
print(random_float_matrix)
```

```python
# random integer matrix (between any two integers)
random_integer_matrix = np.random.randint(5, 100, size = (4, 3))
print(random_integer_matrix)
```

Matrix with all the values as 1

```python
# float
one_matrix_float = np.ones((2, 3))
print(one_matrix_float)
```

```python
# integer
one_matrix_integer = np.ones((2, 3), dtype = int)
print(one_matrix_integer)
```

Null matrix or Zero Matrix

```python
# float
zero_matrix_float = np.zeros((3, 2))
print(zero_matrix_float)
```

```python
# integer
zero_matrix_integer = np.zeros((3, 2), dtype = int)
print(zero_matrix_integer)
```

Identity Matrix (always a square)

only single set of parantheses for identity matrix

```python
# float
identity_matrix_float = np.eye(3, 3)
print(identity_matrix_float)
```

```python
# integer
identity_matrix_integer = np.eye(3, 3, dtype = int)
print(identity_matrix_integer)
```

Transpose of Matrix

```python
matrix = np.array([[2, 3, 4], [5, 6, 7]])
print(matrix)

transpose_matrix = np.transpose(matrix)
print(transpose_matrix)
```
