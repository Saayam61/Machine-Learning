```python
import numpy as np
```

Dot (Scalar) product of two vectors: results in a scalar

```python
vector1 = np.array([2, 3])
vector2 = np.array([4, 4])
dot_product = np.dot(vector1, vector2)
print(dot_product)
```

```python
vector1 = np.array([2, 3, 4])
vector2 = np.array([5, 6, 7])
dot_product2 = np.dot(vector1, vector2)
print(dot_product2)
```

Cross product of two vectors: results in a vector

```python
vector1 = np.array([2, 3])
vector2 = np.array([4, 4])
cross_product = np.cross(vector1, vector2)
print(cross_product) # this -4 is not scalar but a {0; 0; 4} vector
```

```python
vector1 = np.array([2, 3, 6])
vector2 = np.array([4, 4, 5])
cross_product2 = np.cross(vector1, vector2)
print(cross_product2)
```

Projection of "a" vector on "b" vector

proj_b(a) = ((a ⋅ b ) / (b ⋅ b)) * b

or

proj_b(a) = ((a ⋅ b ) / (||b||)^2) * b

```python
a = np.array([2, 5])
b = np.array([8, -6])

# magnitude of "b" vector
b_mag = np.sqrt(sum(b**2))
# or directly use built in function as b_mag = np.linalg.norm(b)


proj_of_a_on_b = ((np.dot(a, b) / b_mag**2) * b)
# b_mag**2 is same as dot product of (b, b). b_mag^2 is np.dot(b, b)

print(proj_of_a_on_b)
```

```python
a = np.array([23, 45, 62])
b = np.array([45, 82, 67])

# magnitude of "b" vector
b_mag = np.sqrt(sum(b**2))
# or directly use built in function as b_mag = np.linalg.norm(b)


proj_of_a_on_b = ((np.dot(a, b) / np.dot(b, b)) * b)
# b_mag**2 is same as dot product of (b, b). b_mag**2 is np.dot(b, b)

print(proj_of_a_on_b)
```
