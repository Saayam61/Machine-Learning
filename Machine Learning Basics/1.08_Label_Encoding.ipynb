Label Encoding (commonly used in classification problems):

- Converting the labels (categorical data) into numeric form

```python
# importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset (2 Labels)

```python
# loading the data from a csv file to a pandas data frame
cancer_df = pd.read_csv('/content/breast_cancer_data.csv')
```

```python
cancer_df.head()
```

```python
# finding the count of different labels
cancer_df['diagnosis'].value_counts()
```

```python
# load the label encoder function
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_df['diagnosis'])
```

```python
# appending the labels to the dataframe
cancer_df['target'] = labels
```

```python
cancer_df.head()
```

0 --> Benign

1 --> Malignant

```python
cancer_df['target'].value_counts()
# since we have not explicitly defined which should be 0 and 1, 0 will be assigned to that which will
# come first alphabetically.
```

Label Encoding of iris dataset (3 labels)

```python
# loading the data from a csv file to a pandas data frame
iris_df = pd.read_csv('/content/iris.csv')
```

```python
iris_df.head()
```

```python
iris_df['Species'].value_counts()
```

```python
# loading the label encoder
label_encoder2 = LabelEncoder()
```

```python
iris_labels = label_encoder2.fit_transform(iris_df['Species'])
```

```python
iris_df['target'] = iris_labels
```

```python
iris_df.head()
```

Iris-setosa --> 0

Iris-versicolor -> 1

Iris-virginica -> 2
