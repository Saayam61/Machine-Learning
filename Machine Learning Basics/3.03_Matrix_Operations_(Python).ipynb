```python
import numpy as np
```

Matrix Addition (Shape of Matrices should be same)

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(A)
print(B)
```

```python
print(A.shape, B.shape)
```

```python
sum = A + B
print(sum)
```

OR

```python
sum = np.add(A, B)
print(sum)
```

Matrix Subtraction (Shape of Matrices should be same)

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(A)
print(B)
```

```python
diff = A - B
print(diff)
```

OR

```python
diff = np.subtract(A, B)
print(diff)
```

Multiplying a Matrix by a Scalar

```python
A = np.array([[1, 2], [3, 4]])
print(A)
```

```python
scalar_multiply = 5 * A
print(scalar_multiply)
```

OR

```python
scalar_multiply = np.multiply(5, A) # np.product is for scalar
print(scalar_multiply)
```

Multiplying 2 Matrices

- the shape Required for matrix multiplication is **(a, b)** and **(b, c)** where the Resultant Matrix will be **(a, c)**

```python
A = np.array([[1, 2], [3, 4], [2, 3]])
B = np.array([[5, 6], [7, 8]])
print(A.shape, B.shape)
```

```python
print(A)
```

```python
print(B)
```

```python
# np.dot: A versatile function that performs different operations based on input dimensions.
# 1D arrays (Vectors): Computes the dot product.
# 2D arrays (Matrices): Performs matrix multiplication.

# or you can use A @ B for matrix multiplication
matrix_product = np.dot(A, B)
print(matrix_product)
```

```python
matrix_product.shape
```
